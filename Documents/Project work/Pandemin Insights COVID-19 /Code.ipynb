```python
# Data Cleaning/ Preprocessing

# Removing unecessary columns from the CovidDeaths dataset
# Kept only the 8 relevant columns

keep = [
    'continent', 'location', 'date', 'total_cases', 'new_cases', 'total_deaths',
    'new_deaths', 'population'
        ]
covid_df = covid_df[keep].copy()
```

```python
covid_df.head(20)
```

```python
# Changing date format 
covid_df['date'] = pd.to_datetime(covid_df['date'], format = '%Y-%M-%D')
```

```python
covid_df.head(15)
```

```python
# Removing null values 
covid_death = covid_df.dropna()
vaccine = vaccine_df.dropna()
```

```python

covid_death.sort_values(
    by= ['location', 'date'],
    ascending= [True, True],
    inplace= True
)
```

```python
# Exploratory Data Analysis

#Calculate death percentage
#covid_deaths['death_percentage'] = covid_deaths['total_deaths']/ covid_deaths['total_cases'] *100

covid_death['death_percentage'] = covid_death['total_deaths'] / covid_death['total_cases']

covid_death.head(10)
```

```python
# Percentage population effected

covid_death['percent_population_infected'] = covid_death['total_cases']/covid_death['population']

covid_death
```

```python
# Country level max infection and infection rate 
# Finding each country's worst outbreak day and then converting that number 
#into a percent of it's population

# Group data by country
max_inf = (
covid_death.groupby(['location', 'population'], as_index= False)
['total_cases'].max()
)
max_inf['max_percent_pop_infected'] = max_inf['total_cases']/ max_inf['population'] * 100

max_inf
```

```python
# Calculating country level max deaths

max_death = (
    covid_death.groupby(['location'], as_index= False)
    ['total_deaths'].max()
    )

max_death

# Top 1 country by max deaths
highest = max_death.nlargest(1, 'total_deaths')
print(highest)

# Bottom 1 country by max deaths
lowest  = max_death.nsmallest(1, 'total_deaths')
print(lowest)


```

```python
# Continent level max infection

import pandas as pd

# 1. Compute peak total_cases per continent
max_con = covid_death.groupby('continent', as_index=False) \
                     .agg(max_cases=('total_cases', 'max'))

# 2. Compute total population per continent
pop_con = covid_death.groupby('continent', as_index=False) \
                     .agg(total_population=('population', 'sum'))

# 3. Merge them together
continent_stats = max_con.merge(pop_con, on='continent')

# 4. Calculate infection rate (%) at the worst point
continent_stats['infection_rate_pct'] = (
    continent_stats['max_cases'] 
    / continent_stats['total_population'] 
    * 100
)

print(continent_stats)


```

```python
# Continent level- max deaths
max_death_con = (
    covid_death.groupby(['continent'], as_index= False)
    ['total_deaths'].max()
    )

max_death_con
```

```python
# Producs a concise, high level view of pandemic's impact total infections, total fatalities
# and overall mortality rate 
# Calculating Total cases, total deaths, and death percentage overall 

global_stats = pd.DataFrame({
    'total_cases': [covid_death['new_cases'].sum()],
    'total_deaths': [covid_death['new_deaths'].sum()]
})
global_stats['death_percentage'] = global_stats['total_deaths'] / global_stats['total_cases'] * 100

global_stats
```

```python
# merge Covid_death and Vaccination Data

vaccine_data = pd.read_excel("CovidVaccinations.xlsx")

vaccine_data
```

```python
# Drop any duplicate (location, date) rows in each

covid_death = covid_death.drop_duplicates(subset=['location', 'date'])
vaccine_data = vaccine_data.drop_duplicates(subset=['location', 'date'])

# Merge on location and date (inner join keeps only macthing rows)

df= pd.merge(covid_death, vaccine, on =['location', 'date'], 
             how = 'inner', 
             suffixes= ('_death', '_vaccine')
             )
```

```python
# Sort for readability and times series operations

df = df.sort_values(['location', 'date']).reset_index(drop= True)
```

```python
# Now df containes all covid death + vaccination columns with no duplicates
df
```

```python
# Test how vaccination pace related to COVID 19 mortality
# We need two core metrics
# 1. Mortality rate ( Daily Deaths per 100k population)
# 2. Vaccination Coverage (e.g. cummulative full vaccinated people/population *100

# Once computed, we can verify
# Overall correlation: Across all country -days, do higher vaccination
# levels coincide with lower mortality rates?
# High Vs low comparison: On days when a country's vaccine coverage was above X%, how did it's average death rate compare
# to days below X%?


# Calculating mortality rate per 100k population 
df['death_rate_per_100k'] = df['new_deaths']/df['population'] * 1000000


# Vaccination coverage %(Fully Vaccinated)
df['vax_coverage_pct'] = (df['people_fully_vaccinated']/ df['population'] * 100)
```

```python
# Overall correlation

# Pearson Correlation

corr = df[['vax_coverage_pct', 'death_rate_per_100k']].corr().iloc[0,1]
print(f"Correlation between vaccine coverage and death rate: {corr:.2f}")
```
